```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
mu=100
sigma=50
x= mu + sigma* np.random.randn(1000)


plt.figure(figsize=(25,10) , facecolor='orange')
b=plt.axes()
b.set_facecolor('pink')
plt.xlabel("x-asise")
plt.ylabel("y-asise")
plt.title("histogram")
plt.grid(color="m" ,ls= '--',linewidth=0.5)

n, bins , _ = plt.hist(x,bins=20 ,density=1, facecolor='green',alpha=0.7 )

y= ( ( 1/( np.sqrt( 2*np.pi * sigma**2 ) ) ) ) * np.exp(-0.5 * (1/sigma * (bins - mu)) **2)

plt.scatter(bins,y,color="red",s=50,marker='o')

plt.plot(bins ,y, 'y--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

x=pd.read_csv(r"C:\Users\ishuk\Desktop\intership\jupyters_lab\pandess\iris2.csv" , names =['s1' ,'sw' , 'pl', 'pw' , 'class'])

plt.figure(figsize=(20,6) , facecolor='orange')
b=plt.axes()
b.set_facecolor('pink')
plt.xlabel("x-asise")
plt.ylabel("y-asise")
plt.title("histogram")
plt.grid(color="m" ,ls= '--',linewidth=0.5)

sns.kdeplot(x.loc[(x['class']=='Iris-setosa'), 's1'],
            x.loc[(x['class']=='Iris-setosa'), 'sw'],
            color='green', shade =True ,shade_lowest=False)

#.........................................................................................................................

plt.figure(figsize=(25,6) , facecolor='orange')

plt.subplot(1,4,1)
plt.title("histogram")
plt.grid(color="m" ,ls= '--',linewidth=0.5)
sns.kdeplot(x.loc[(x['class']=='Iris-setosa'), 's1'],
            x.loc[(x['class']=='Iris-setosa'), 'sw'],
            color='green', shade =True ,shade_lowest=True)

plt.subplot(1,4,2)
plt.title("histogram")
plt.grid(color="m" ,ls= '--',linewidth=0.5)
sns.kdeplot(x.loc[(x['class']=='Iris-setosa'), 'pl'],
            x.loc[(x['class']=='Iris-setosa'), 'sw'],
            color='yellow', shade =True ,shade_lowest=False)

plt.subplot(1,4,3)
plt.title("histogram")
plt.grid(color="m" ,ls= '--',linewidth=0.5)
sns.kdeplot(x.loc[(x['class']=='Iris-setosa'), 's1'],
            x.loc[(x['class']=='Iris-setosa'), 'pl'],
            color='red', shade =False ,shade_lowest=True,alpha=1)

plt.subplot(1,4,4)
plt.title("histogram")
plt.grid(color="m" ,ls= '--',linewidth=0.5)
sns.kdeplot(x.loc[(x['class']=='Iris-setosa'), 'pw'],
            x.loc[(x['class']=='Iris-setosa'), 's1'],
            color='blue', shade =False ,shade_lowest=False,alpha=1)

```
