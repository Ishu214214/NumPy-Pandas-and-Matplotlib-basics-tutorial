```python
bike=['Rajdutte','Ninja','Apachi','KTM', 'Yahamaha','Hero Honda ']
price=[40,30,44,112,34,56]

plt.figure(figsize=(20,10) , facecolor='pink')
#plt.subplot(1,8,1)
plt.pie( price , labels=bike , startangle=45 )
plt.show()

plt.figure(figsize=(20,10) , facecolor='pink')
#plt.subplot(1,8,2)
plt.pie( price , labels=bike , startangle=45 , explode=[0.2,0,0,0,0,0] )
plt.show()

plt.figure(figsize=(20,10) , facecolor='pink')
#plt.subplot(1,8,3)
plt.pie( price , labels=bike , startangle=45 , explode=[0.2,0,0,0,0,0] , shadow=True )
plt.show()

#plt.subplot(1,8,4)
plt.figure(figsize=(20,10) , facecolor='pink')
plt.pie( price , labels=bike , startangle=45 , explode=[0.2,0,0,0,0,0] , shadow=True ,wedgeprops={ "linewidth" : 3, "edgecolor":"black" } )
plt.show()

#plt.subplot(1,8,5)
plt.figure(figsize=(20,10) , facecolor='pink')
plt.pie( price , labels=bike , startangle=45 , explode=[0.2,0,0,0,0,0] , shadow=True ,wedgeprops={ "linewidth" : 3, "edgecolor":"black" } , autopct=str(price) )
plt.show()

#plt.subplot(1,8,6)
plt.figure(figsize=(20,10) , facecolor='pink')
plt.pie( price , labels=bike , startangle=45 , explode=[0.2,0,0,0,0,0] , shadow=True ,wedgeprops={ "linewidth" : 3, "edgecolor":"black" } , autopct='%.2f' )
plt.show()
```
